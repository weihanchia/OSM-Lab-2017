```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 8.1

```python
# Plot Feasible Set
x = np.linspace(0,6,200)
y1 = 2./3*x + 4./3 # >
y2 = 1./6*x - 1./6 # >
y3 = 6 - x # <

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
ymax = y3
ymin = np.maximum(y1, y2)
plt.fill_between(x, ymax, ymin, where=ymax>ymin, color='grey', alpha=0.5)
plt.show()
```

```python
# Plot Objective Function over Feasible Set

x = np.linspace(0,2.8, 50)
y = np.linspace(4/.3, 3.2, 50)
xgrid, ygrid = np.meshgrid(x, y)
z = 5*xgrid - 4*ygrid

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z, cmap="magma")
plt.show()
```

#### 8.2

```python
# Plot Feasible Set
x = np.linspace(0,15,200)
y1 = 15 - x # <
y2 = 6 - 2./3*x # <
y3 = x - 4 # >

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
ymax = np.minimum(y1, y2)
ymin = np.maximum(y3, 0)
plt.fill_between(x, ymax, ymin, where=ymax>ymin, color='grey', alpha=0.5)
plt.show()
```

```python
# Evaluating Function at each vertex.

def f1(x):
    return 3*x[0] + x[1]

print(f1((0,0)))
print(f1((0,6)))
print(f1((4,0)))
print(f1((6,2)))
```

```python
# Plot Feasible Set
x = np.linspace(0,30,200)
y1 = 11 + x # <
y2 = 27 - x # <
y3 = 90./5 - 2./5*x # <

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
ymax = np.minimum(y1, np.minimum(y2,y3))
ymin = 0
plt.fill_between(x, ymax, ymin, where=ymax>ymin, color='grey', alpha=0.5)
plt.show()
```
